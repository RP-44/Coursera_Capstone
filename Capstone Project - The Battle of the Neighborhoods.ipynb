### Introduction: Business Problem

#### Background 
 
With the globalization of food, people have tended to want to try new cuisines. Portuguese cuisine, despite influencing other cuisines, due to colonization, is little known.

As such, my client wants to open a typical Portuguese food restaurant in Chicago, IL.

The objective of this project is to find the best localization to open a Portuguese restaurant in Chicago, IL.
 
#### Problem 
In order for the business to be successful, i need to find the best localization.
For that purpose i will gather data about other restaurant localization, competitors(Portuguese restaurant) and top venues.

### Data
